```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import analyses.util as util
import pandas as pd
import numpy as np
import qgrid

import analyses.model_belief_fit as mbf
```

```python
matplotlib.rc('figure', facecolor=(1, 1, 1, 1))
```

```python
model = mbf.load("results")
```

```python
laplace_prior = mbf.make_prior(mbf.log_laplace, mu=1, b=1)
```

```python
res_G = model\
    .groupby(level='version')\
    .get_group('G')\
    .groupby(level=['model', 'pid'])\
    .apply(mbf.fit_responses, laplace_prior)
params_G = res_G\
    .reset_index()[['model', 'pid', 'B']]\
    .drop_duplicates()\
    .set_index(['model', 'pid'])['B']
```

```python
empirical_priors = params_G.groupby(level='model').apply(lambda x: mbf.make_prior(mbf.kde, np.asarray(x), 0.1))
```

```python
from IPython.html.widgets import interact
```

```python
keys = sorted(model.groupby(level='version').get_group('I').index.get_level_values('pid').unique())
```

```python
@interact(i=(0, len(keys)-1))
def plot(i):
    fig, axes = plt.subplots(1, 3)
    for j, (model_name, df) in enumerate(model.groupby(level='version').get_group('I').groupby(level='model')):
        ax = axes[j]
        df2 = df.groupby(level='pid').get_group(keys[i])
        y = np.asarray(df2['responses'])
        X = np.asarray(df2['llr'])

        B0 = mbf.logistic_regression(X, y, laplace_prior, verbose=False)
        post = mbf.make_posterior(X, y, laplace_prior)
        B = np.linspace(-3.0, 3.0, 200)
        pB = np.array([post(x) for x in B])
        ax.plot(B, pB, color='b')
        ax.set_ylim(-1, 25)
        ax.vlines(B0, *ax.get_ylim(), color='b', linestyle='--')

        B0 = mbf.logistic_regression(X, y, empirical_priors[model_name], verbose=False)
        post = mbf.make_posterior(X, y, empirical_priors[model_name])
        B = np.linspace(-3.0, 3.0, 200)
        pB = np.array([post(x) for x in B])
        ax.plot(B, pB, color='r')
        ax.vlines(B0, *ax.get_ylim(), color='r', linestyle='--')

        ax.set_title(model_name)

    fig.set_figwidth(14)
```

```python
data = pd.read_csv("results/model_belief_fit.csv").set_index(['model', 'likelihood', 'version'])
```

```python
def plot(ax, df, legend=False):
    human_avg = df['correct'].groupby(level='trial').mean()
    ax.plot(human_avg.index, human_avg, label="human", lw=2)

    model_avg = df['p correct'].unstack('model').groupby(level='trial').mean()
    for model in model_avg:
        ax.plot(model_avg[model].index, model_avg[model], label=model, lw=2)
    
    if legend:
        ax.legend(loc='best')
    ax.set_ylim(0.45, 1)
```

```python
def plot_hist(ax, params, legend=False):
    colors = ['r', 'g', 'b']

    for i, (m, df) in enumerate(params.groupby(level='model')):
        x = np.asarray(df['B'].dropna())
        ax.hist(x, label=m, bins=32, alpha=0.5, range=[-0.5, 2], normed=True, color=colors[i])
    
        x = np.linspace(-0.5, 2.0, 100)
        y = mbf.kde(x[:, None], np.asarray(df['B'].dropna()), 0.1)
        ax.plot(x, np.exp(y), color=colors[i])

    if legend:
        ax.legend(loc='best')
```

```python
fig, axes = plt.subplots(1, 3, sharey=True)
order = ['G', 'H', 'I']
for version, df in data.groupby(level='version'):
    i = order.index(version)
    plot_hist(axes[i], df, legend=(i == (len(order) - 1)))
    axes[i].set_title(version)
fig.set_figwidth(14)
fig.set_figheight(3)
plt.tight_layout()
```
