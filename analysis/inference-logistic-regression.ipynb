```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import analyses.util as util
import pandas as pd
import numpy as np
import qgrid
from scipy.optimize import minimize
```

```python
data = util.load_all()
version = 'G'
```

```python
responses = data['human']['C']\
    .set_index(['version', 'pid', 'trial', 'stimulus'])['mass? response']\
    .dropna()\
    .sortlevel()
responses = (responses + 1) / 2
```

```python
correct = data['human']['C']\
    .set_index(['version', 'pid', 'trial', 'stimulus'])['mass? correct']\
    .dropna()\
    .sortlevel()
```

```python
belief = pd.read_csv("results/model_belief_by_trial.csv")
```

```python
empirical = belief\
    .groupby(('likelihood', 'query'))\
    .get_group(('empirical', 'fall cf'))\
    .drop(['sigma', 'phi'], axis=1)
```

```python
llh = empirical\
    .set_index(['version', 'model', 'pid', 'trial', 'stimulus', 'hypothesis'])['logp']\
    .unstack('hypothesis')\
    .sortlevel()
```

```python
llr = (llh[1] - llh[-1]).unstack('model').ix[responses.index]
```

```python
def log_laplace(x, mu=0, b=1):
    # (1 / 2*b) * np.exp(-np.abs(x - mu) / b)
    c = -np.log(2 * b)
    e = -np.abs(x - mu) / b
    return c + e
```

```python
def kde(x, params, sigma):
    p = -0.5 * np.log(2 * np.pi * sigma**2) - (x - params) ** 2 / (2 * sigma ** 2)
    logp = np.log(np.mean(np.exp(p), axis=-1))
    return logp
```

```python
def make_prior(f, *args, **kwargs):
    def prior(x):
        return f(x, *args, **kwargs)
    return prior
```

```python
def logistic_regression(X, y, prior_func):
    def log_posterior(B):
        p = 1.0 / (1 + np.exp(-(X * B)))
        log_lh = np.log((y * p) + ((1 - y) * (1 - p))).sum()
        log_prior = prior_func(B)
        log_posterior = log_lh + log_prior
        return -log_posterior
    
    res = minimize(fun=log_posterior, x0=1.0)
    return float(res['x'])
```

```python
def fit_responses(df, prior_func):
    y = np.asarray(df['responses'])
    X = np.asarray(df['llr'])
    
    if 'chance' in df.name:
        B = 0
    else:
        B = logistic_regression(X, y, prior_func)
        
    f = X * B
    mu = 1.0 / (1 + np.exp(-f))
    
    mask = np.asarray(df['responses'] != df['correct'])
    mu[mask] = 1 - mu[mask]
    
    new_df = df.copy()
    new_df['p correct'] = mu
    new_df['f'] = f
    new_df['B'] = B
    return new_df
```

```python
def plot(ax, df, legend=False):
    human_avg = df['correct'].groupby(level='trial').mean()
    ax.plot(human_avg.index, human_avg, label="human", lw=2)

    model_avg = df['p correct'].unstack('model').groupby(level='trial').mean()
    for model in model_avg:
        ax.plot(model_avg[model].index, model_avg[model], label=model, lw=2)
    
    if legend:
        ax.legend(loc='best')
    ax.set_ylim(0.45, 1)
```

```python
def plot_hist(ax, params, legend=False):
    colors = ['r', 'g', 'b']

    for i, (m, df) in enumerate(params.groupby(level='model')):
        x = np.asarray(df['B'].dropna())
        ax.hist(x, label=m, bins=32, alpha=0.5, range=[-0.5, 2], normed=True, color=colors[i])
    
        x = np.linspace(-0.5, 2.0, 100)
        y = kde(x[:, None], np.asarray(df['B'].dropna()), 0.1)
        ax.plot(x, np.exp(y), color=colors[i])

    if legend:
        ax.legend(loc='best')
```

```python
model = llr.copy()
model['responses'] = responses
model['correct'] = correct
model = model.set_index(['responses', 'correct'], append=True)\
    .stack()\
    .reset_index(['responses', 'correct'])\
    .rename(columns={0: 'llr'})
model.index.names = ['version', 'pid', 'trial', 'stimulus', 'model']
model = model.reorder_levels(['version', 'model', 'pid', 'trial', 'stimulus']).sortlevel()
```

Fit all responses to global parameter settings:

```python
laplace_prior = make_prior(log_laplace, mu=1, b=1)
res_all = model.groupby(level=['version', 'model']).apply(fit_responses, laplace_prior)
params_all = res_all.reset_index()[['version', 'model', 'B']].drop_duplicates().set_index(['version', 'model'])
params_all
```
