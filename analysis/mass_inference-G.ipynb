## Experimental questions

2. Are people running simulations to compute the likelihood term?

3. Can people integrate information over multiple trials? Are they
updating their beliefs according to Bayes' Rule?

```python
def get_fall(arr):
    return arr.pivot_table(
	rows=['kappa0', 'pid'],
	cols='stimulus',
	values='fall? response')

def get_mass(arr):
    return arr.pivot_table(
	rows=['kappa0', 'pid'],
	cols='trial',
	values='mass? response')

def get_trials(arr):
    return arr.pivot(
	index='trial', 
	columns='pid', 
	values='stimulus')
```

```python
exp = {
    'A': exp_all.groupby('mode').get_group('experimentA'),
    'B': exp_all.groupby('mode').get_group('experimentB'),
    'C': exp_all.groupby('mode').get_group('experimentC')
}
```

```python
trials = {mode: get_trials(exp[mode]).dropna() for mode in exp}
```

```python
ipe = {
    'A': IPE("mass_inference-G-a_ipe_fall", sigma=0.04, phi=0.2),
    'B': IPE("mass_inference-G-b_ipe_fall", sigma=0.04, phi=0.2),
    'C': IPE("mass_inference-G-b_ipe_fall", sigma=0.04, phi=0.2)
}

fb = {
    'A': NoFeedback("mass_inference-G-a_truth_fall"),
    'B': NoFeedback("mass_inference-G-b_truth_fall"),
    'C': Feedback("mass_inference-G-b_truth_fall"),
}
```

```python
kappas = (-1.0, 1.0)
```

```python
def make_accurate_prior(ipe, kappa0, kappas):
    prior = np.zeros(len(kappas))
    prior[kappas.index(kappa0)] = 1
    return prior
```

```python
def gen_model_fall(arr, mode):
    kappa0, pid = arr.name
    model = ObserverModel(
	ipe[mode], fb[mode], trials[mode][pid], kappa0,
	kappas=kappas,
	prior=make_accurate_prior(ipe[mode], kappa0, kappas))
    J_human = arr.reset_index(drop=True).stack().ix[0].ix[trials[mode][pid]]
    J_model = pd.Series(np.exp(model.p_fall), index=J_human.index)
    J = pd.DataFrame({
	'human': J_human,
	'model': J_model
    })
    J.index.name = "stimulus"
    J.columns.name = "species"
    return J

def fall_stats(df):
    stats = df.stack('species')\
	      .groupby(level=['kappa0', 'stimulus', 'species'])\
	      .apply(at.bootstrap)\
	      .unstack('species')\
	      .unstack()
    return stats
```

```python
A_J_fall = get_fall(exp['A'])\
    .groupby(level=["kappa0", "pid"])\
    .apply(gen_model_fall, 'A')
```

```python
B_J_fall = get_fall(exp['B'])\
    .groupby(level=["kappa0", "pid"])\
    .apply(gen_model_fall, 'B')
```

```python
A_stats = fall_stats(A_J_fall)
B_stats = fall_stats(B_J_fall)
```

```python
def plot_fall(arr):
    fig, ax = plt.subplots()
    for key, group in arr.groupby(level='kappa0'):
	hm = group['human', 'median']
	hl = hm - group['human', 'lower']
	hu = group['human', 'upper'] - hm
	mm = group['model', 'median']
	ml = mm - group['model', 'lower']
	mu = group['model', 'upper'] - mm
	ax.errorbar(hm, mm, xerr=[hl, hu], yerr=[ml, mu], label=key, ls='', marker='o')
    ax.set_xlim(1, 7)
    ax.set_ylim(0, 1)
    ax.set_xlabel("Human")
    ax.set_ylabel("Model")
    ax.legend(loc=0, title='kappa0')
```

```python
print A_stats.xs('median', level=1, axis=1).unstack('kappa0').corr()
plot_fall(A_stats)
```

```python
print B_stats.xs('median', level=1, axis=1).unstack('kappa0').corr()
plot_fall(B_stats)
```

```python
def find_switchpoint(df):
    kappa0, pid = df.name
    arr = np.asarray(df).copy().squeeze()

    last = arr[-1]
    if (kappa0 < 0 and last == 1) or (kappa0 > 0 and last == 0):
	new_arr = np.zeros(arr.shape)
    else:
	eq = np.nonzero(arr != last)[0]
	if eq.size == 0:
	    idx = 0
	elif eq[-1] == (arr.size - 2):
	    idx = arr.size
	else:
	    idx = eq[-1] + 1

	new_arr = np.empty(arr.shape)
	new_arr[:idx] = False
	new_arr[idx:] = True

    new_df = pd.DataFrame(new_arr[None], index=df.index, columns=df.columns)
    return new_df
```

```python
def gen_model_mass(arr, mode):
    kappa0, pid = arr.name
    apriori_model = ObserverModel(
	ipe[mode], fb[mode], trials[mode][pid], kappa0, kappas=kappas)
    empirical_model = ObserverModel(
	EmpiricalIPE(exp['B']), fb[mode], trials[mode][pid], kappa0, kappas=kappas)

    J_human = arr.reset_index(drop=True)
    J_human = J_human.stack().ix[0]
    idx = np.array(J_human.index)

    J_apriori_model = pd.Series(
	np.exp(apriori_model.p_mass[idx]), index=idx)
    J_empirical_model = pd.Series(
	np.exp(empirical_model.p_mass[idx]), index=idx)

    J = pd.DataFrame({
	'human': J_human,
	'apriori_model': J_apriori_model,
	'empirical_model': J_empirical_model,
    })
    J.index.name = "stimulus"
    J.columns.name = "species"
    return J

def mass_stats(df):
    stats = df\
	.unstack()\
	.groupby(level='kappa0')\
	.agg([at.binom_conf_lower, at.binom_mean, at.binom_conf_upper])\
	.stack(0)\
	.stack(0)\
    	.rename(columns={
    	    'binom_conf_lower': 'lower',
    	    'binom_mean': 'median',
    	    'binom_conf_upper': 'upper'
    	})
    stats.index.names = ['kappa0', 'species', 'trial']
    return stats
```

```python
C_J_mass = get_mass(exp['C'])\
    .groupby(level=["kappa0", "pid"])\
    .apply(gen_model_mass, 'C')

C_J_mass['switch'] = C_J_mass['human']\
    .unstack('stimulus')\
    .groupby(level=['kappa0', 'pid'])\
    .apply(find_switchpoint)\
    .stack()
```

```python
C_stats = mass_stats(C_J_mass)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, sharey=False)
colors = {
    -1.0: 'r', 
    1.0: 'b'
}

groups = C_stats.groupby(level=['kappa0', 'species'])
for (kappa0, species), df in groups:
    X = df.reset_index(['kappa0', 'species'], drop=True).index
    lower = np.asarray(df['lower'])
    upper = np.asarray(df['upper'])
    median = np.asarray(df['median'])

    if species == 'human':
	ls = "-"
	ax = ax2
    elif species == 'apriori_model':
	ls = "--"
	ax = ax1
    elif species == 'empirical_model':
	ls = "-"
	ax = ax1
    elif species == 'switch':
	ls = "-"
	ax = ax3
    else:
	raise ValueError("unhandled species: %s" % species)

    ax.fill_between(X, lower, upper, alpha=0.2, color=colors[kappa0])
    ax.plot(X, median, label=(kappa0, species), 
	    lw=2, ls=ls, color=colors[kappa0])

    ax.set_xlabel("Trial")
    ax.set_ylim(0, 1)

ax1.set_xlim(1, 20)
ax1.set_xticks([1, 2, 3, 4, 6, 9, 14, 20])
ax1.set_ylabel("<-- color 0 heavy     color 1 heavy -->")
ax1.set_title("Model belief")
ax1.legend(loc=0)

ax2.set_title("Human responses")
ax2.legend(loc=0)

ax3.set_ylabel("% participants")
ax3.set_title("Fraction who determined mass by trial $t$")

fig.set_figwidth(15)
fig.set_figheight(5)
```
