## Load Data

```python
mexp_dict = {}
mbelief_dict = {}

conds = condinfo.groupby(level=CONDITION)
for cond, arr in conds:
    arr.drop_duplicates(inplace=True)
    arr.name = cond
    mexp, mbelief = at.simulate_model(arr, ipe, fb, n_fake=2000)
    mexp_dict[cond] = mexp
    mbelief_dict[cond] = mbelief

mexp = pd.concat(mexp_dict, names=CONDITION).reset_index(-1, drop=True).reset_index()
mbelief = pd.concat(mbelief_dict, names=CONDITION).reset_index(-1, drop=True).reset_index()
```

## Model Analysis
### Plot of smoothed model outcome probabilities

```python
ipe_lh = pd.DataFrame({
    'pfell_mean': (
	ipe
	.groupby(level=PARAMS)
	.agg(at.binom_mean)
	.stack()
    ),
    'pfell_std': (
	ipe
	.groupby(level=PARAMS)
	.agg(at.binom_std)
	.stack()
    ),
    'pfell_smoothed': (
	ipe
	.stack('stimulus').unstack('sample')
	.groupby(level=['sigma', 'phi', 'stimulus'], group_keys=False)
	.apply(at.pfell_smoothed)
	.sortlevel()
    )
})
```

```python
def plot_smoothing(lh, ax, colors):
    stim = lh.name
    x = lh.unstack().reset_index('kappa')
    color = colors[stim]
    line = ax.plot(
        x.kappa, x.pfell_smoothed,
        lw=3, color=color, label=stim)
    points = ax.errorbar(
        x.kappa, x.pfell_mean, x.pfell_std,
        ls='', fmt='o', ms=5, color=color)
    return line, points
```

```python
plotstims = ["tower_00357_0111011000", "tower_00388_0011101001"]
colors = at.stim_colors(plotstims)
fig, ax = plots.smoothing_before()
(ipe_lh
 .stack()
 .unstack('stimulus')[plotstims]
 .apply(plot_smoothing, args=[ax, colors]))
plots.smoothing_after(fig, ax)
savefig("images/likelihood_smoothing", close=False, ext=['png', 'pdf', 'svg'])
```

### Plot of learning model beliefs over time

```python
def plot_belief(arr, ax, max_belief):
    belief = np.exp(arr.pivot('kappa', 'trial', 'log_p'))
    cmap = at.make_cmap("lh", (0, 0, 0), (.55, .55, .55), (1, 1, 1))
    im = ax.imshow(
	belief, cmap=cmap,
	interpolation='nearest',
	vmin=0, vmax=max_belief)
    order, fbtype, ratio = arr.name
    ax.set_title(r"$r_0=%s$, order %s" % (ratio, order))
    return im
```

```python
fb_belief = mbelief.groupby('feedback').get_group('fb')
max_belief = np.exp(fb_belief['log_p']).max()
fig, grid = plots.belief_before(fb_belief)
for ax, (name, group) in izip(grid, fb_belief.groupby(CONDITION)):
    group.name = name
    im = plot_belief(group, ax, max_belief)
plots.belief_after(fig, grid, im)
savefig("images/ideal_observer_beliefs", close=False, ext=['png', 'pdf', 'svg'])
```

## Explicit mass judgments

```python
mass_correct = (
    pd
    .concat([at.get_correct_mass_responses(hexp, 'human'),
	     at.get_correct_mass_responses(mexp, 'model')])
    .set_index(SPECIES + ['pid', 'trial'])
    .correct
    .astype('f8')
    .unstack('trial')
    .dropna(axis=1))
```

```python
pd.options.display.float_format = lambda x: ' %.2f' % x
```

### Chi-square analysis of explicit judgments over time

Is accuracy independent of time?

```python
mass_correct.groupby(level=SPECIES).apply(at.chi2)
```

### Binomial test for above-chance accuracy

Is accuracy above chance?

```python
# P(probability of correct response <= 0.5)
mass_correct.groupby(level=SPECIES).agg(at.test_gt_chance)
```

### Binomial test for same-as-model accuracy

Is human accuracy different from model accuracy?

```python
# P(probability of correct human response != probability of correct model response)
mass_correct.groupby(level=CONDITION).agg(at.test_diff_model)
```

### Binomial test for better-than-model accuracy

Is human accuracy better than model accuracy?

```python
# P(probability of correct human response <= probability of correct model response)
mass_correct.groupby(level=CONDITION).agg(at.test_gt_model)
```

### Binomial test for worse-than-model accuracy

Is human accuracy worse than model accuracy?

```python
# P(probability of correct human response >= probability of correct model response)
mass_correct.groupby(level=CONDITION).agg(at.test_lt_model)
```

### Test for difference in accuracy for binary vs. visual feedback

Is accuracy different between binary and visual feedback conditions?

```python
# P(probability of correct human response >= probability of correct model response)
(mass_correct.groupby(level=['order', 'ratio', 'species'])
 .agg(at.test_vis_gt_bin)
 .xs('human', level='species'))
```

### Learning curves

```python
mass_stats = mass_correct.groupby(level=SPECIES).agg([
    np.mean, at.binom_conf_lower, at.binom_conf_upper])
mass_stats.columns.names = ['trial', None]
mass_stats = mass_stats.stack('trial').reset_index()
```

```python
def plot_learning_curve(arr, ax):
    order, feedback, ratio, species = arr.name

    opts = {}
    opts['lw'] = 4
    opts['color'] = 'k'

    if species == 'human' and feedback == 'fb':
	opts['ls'] = ':'
	opts['label'] = "Binary feedback"
    elif species == 'human' and feedback == 'vfb':
	opts['ls'] = '--'
	opts['label'] = "Visual feedback"
    elif species == 'model' and feedback == 'fb':
	opts['ls'] = '-'
	opts['label'] = "Learning IPE"
    else:
	return

    x = np.asarray(arr.trial)
    mean = np.asarray(arr['mean'])
    lower = np.asarray(arr.binom_conf_lower)
    upper = np.asarray(arr.binom_conf_upper)

    ax.fill_between(x, lower, upper, color=opts['color'], alpha=0.25)
    ax.plot(x, mean, **opts)
```

```python
curves = mass_stats.groupby(['order', 'feedback', 'ratio', 'species'])
fig, axes = plots.learning_curves_before(mass_stats)
for name, curve in curves:
    curve.name = name
    order, feedback, ratio, species = name
    plot_learning_curve(curve, axes[(order, ratio)])
plots.learning_curves_after(fig, axes)
savefig("images/explicit_mass_judgments", close=False, ext=['png', 'pdf', 'svg'])
savefig("../figures/fig3-explicit_mass_judgments", close=False, ext=['pdf'])
```

## Model Comparison
### Table of model likelihoods

```python
hgroups = hexp.groupby(CONDITION)
trial_lh = (
    pd.concat([hgroups.apply(at.evaluate_model, 'chance', ipe, fb),
	       hgroups.apply(at.evaluate_model, 'static', ipe, fb),
	       hgroups.apply(at.evaluate_model, 'learning', ipe, fb)], 
	      ignore_index=True)
    .pivot_table(
	rows=CONDITION + ['model', 'trial', 'stimulus'],
	cols='pid',
	values='lh'))
trial_lh
```

```python
pid_lh = (trial_lh
	  .groupby(level=CONDITION + ['model'])
	  .sum()
	  .stack())
pid_lh
```

```python
model_lh = (pid_lh
	    .groupby(level=CONDITION + ['model'])
	    .apply(at.bootstrap, rso)
	    .unstack())
model_lh
```

```python
model_lh['median'].unstack('model')
```

```python
# probability that learning model has a higher likelihood than the
# static model
sig_lh = (pid_lh
	  .groupby(level=CONDITION)
	  .apply(at.bootstrap_test_gt, 'static', 'learning', rso))
sig_lh
```

### Plots of likelihoods

```python
plot_lh = (model_lh
	   .drop(['nfb'], level='feedback')
	   .sortlevel()
	   .reset_index('model')
	   .reorder_levels(['order', 'ratio', 'feedback']))

chance = np.asarray(plot_lh.groupby('model').get_group('chance')['median'])
def sub(x):
    y = x.copy()
    y['lerr'] = x['median'] - x['lower']
    y['uerr'] = x['upper'] - x['median']
    y['relative median'] = x['median'] - chance
    return y

plot_lh = (plot_lh
	   .set_index('model', append=True)
	   .groupby(level='model')
	   .apply(sub))
```

```python
def plot_model_comparison(lh, ax):
    colors = {
	'chance': 'k',
	'static': 'm',
	'learning': '#FFA500'
    }
    labels = {
	'chance': 'Chance',
	'static': 'Fixed belief IPE',
	'learning': 'Learning IPE'
    }
    models = ['static', 'learning']

    if lh.name == 'chance':
	chance = float(np.round(lh['median'], decimals=6).unique())
	ax.plot(
	    ax.get_xlim(), [chance]*2, 
	    ls='-', 
	    lw=3,
	    color=colors[lh.name], 
	    label=labels[lh.name])
    elif lh.name in models:
	width = 2./(3.*len(models))
	i = models.index(lh.name)
	x = np.arange(lh.shape[0]) + 1./6.
	ax.bar(
	    x+i*width, lh['relative median'], 
	    bottom=lh['median'] - lh['relative median'],
	    yerr=[lh['lerr'], lh['uerr']],
	    width=width,
	    ecolor='k',
	    color=colors[lh.name], 
	    label=labels[lh.name])
    else:
	raise ValueError("unhandled model: %s" % lh.name)
```

```python
fig, ax = plots.model_comparison_before(plot_lh, 1-sig_lh)
groups = plot_lh.groupby(level='model')
for name in ['chance', 'static', 'learning']:
    arr = groups.get_group(name)
    arr.name = name
    plot_model_comparison(arr, ax)
plots.model_comparison_after(fig, ax)
savefig("images/model_performance", close=False, ext=['png', 'pdf', 'svg'])
savefig("../figures/fig4-model_performance", close=False, ext=['pdf'])
```
