```python
# Builtins
import shutil
from itertools import izip
# External
import numpy as np
import pandas as pd
import matplotlib
from path import path
# Local
import analysis_tools as at
from snippets import datapackage as dpkg
from style_towers import apply_style
```

```python
matplotlib.rc('font', size=20)
matplotlib.rc('axes', titlesize=24, labelsize=20)
matplotlib.rc('xtick', labelsize=16)
matplotlib.rc('ytick', labelsize=16)
matplotlib.rc('legend', fontsize=20)
matplotlib.rc('figure', figsize=(8, 6))
```

```python
DATAPATH = path("../data")
PARAMS = ['sigma', 'phi', 'kappa']
STIMULUS = ['base', 'bitstr']
```

```python
rso = np.random.RandomState(24890234)
```

```python
# load the simulation data package
model_dp = dpkg.DataPackage.load(
    DATAPATH.joinpath("model/mass_all_fall.dpkg"))
```

```python
def split_stimnames(arr):
    arr['base'] = ["_".join(x.split("_")[:2]) for x in arr['stimulus']]
    arr['bitstr'] = [x.split("_")[2] for x in arr['stimulus']]
    arr = arr.drop('stimulus', axis=1)
    return arr
```

```python
# extract model data, and split stimuli names into base and bit string
model = split_stimnames(model_dp.load_resource('model.csv'))
mresp = model.pivot_table(
    rows=PARAMS + ['base', 'bitstr'],
    cols='sample',
    values='nfell')
```

```python
def format_model(arr):
    return arr\
	.reset_index(['sigma', 'phi'], drop=True)\
	.reorder_levels([1, 2, 0])\
	.sortlevel()

def make_fb(arr):
    mean = arr.mean(axis=1)
    fb = format_model(mean.apply(
	lambda x: 1 if x > 2 else 0 if x == 0 else np.nan))\
	.unstack('kappa')\
	.dropna()\
	.stack('kappa')
    fb.index.names[-1] = 'kappa0'
    return fb

def make_ipe(arr):
    return format_model(arr > 4)
```

```python
def savestims(stims, old_dir, new_dir, style):

    # create new directory for stimuli
    old_dir = path("../resources/sso/" + old_dir)
    new_dir = path("../resources/sso/" + new_dir)
    if new_dir.exists():
	shutil.rmtree(new_dir)
    new_dir.makedirs_p()

    # copy stimuli from old directory to new directory
    old_paths = [old_dir.joinpath("_".join(x) + ".cpo") for x in stims.index]
    new_paths = [new_dir.joinpath("_".join(x) + ".cpo") for x in stims.index]
    for old_path, new_path in zip(old_paths, new_paths):
	shutil.copy(old_path, new_path)

    # apply graphics and physics style to new stimuli
    apply_style(new_paths, style, rso=rso)
```

```python
# extract feedback and ipe samples
sigmas = mresp.groupby(level='sigma')
fb = make_fb(sigmas.get_group(0))
ipe = make_ipe(sigmas.get_group(0.05))
```

```python
# compute IPE p(fall | ratio)
pfell = ipe.groupby(level=STIMULUS, group_keys=False)\
	   .apply(at.pfell_smoothed)\
	   .sortlevel()
```

```python
# combine the likelihood function and feedback into one dataframe
lh = pd.merge(
    pfell.reset_index().rename(columns={0: 'P(fall|kappa)'}),
    fb.reset_index().rename(columns={0: 'data'}),
    on=STIMULUS)

lh = lh.set_index(STIMULUS + ['kappa0', 'kappa']).sortlevel()
lh.head()
```

```python
# compute the likelihood of the observed data
lh['P(data|kappa)'] = (
    (lh['P(fall|kappa)'] * lh['data']) +
    ((1 - lh['P(fall|kappa)']) * (1 - lh['data'])))
lh.head()
```

```python
# create a new dataframe with likelihoods for true mass ratios and
# hypothesized ratios
lhr = pd.DataFrame({
    'P(data|kappa0=1,kappa=1)': lh['P(data|kappa)'].ix[:, :, 1.0, 1.0],
    'P(data|kappa0=1,kappa=-1)': lh['P(data|kappa)'].ix[:, :, 1.0, -1.0],
    'P(data|kappa0=-1,kappa=1)': lh['P(data|kappa)'].ix[:, :, -1.0, 1.0],
    'P(data|kappa0=-1,kappa=-1)': lh['P(data|kappa)'].ix[:, :, -1.0, -1.0],
    'P(data|kappa0=1,kappa=0.3)': lh['P(data|kappa)'].ix[:, :, 1.0, 0.3],
    'P(data|kappa0=1,kappa=-0.3)': lh['P(data|kappa)'].ix[:, :, 1.0, -0.3],
    'P(data|kappa0=-1,kappa=0.3)': lh['P(data|kappa)'].ix[:, :, -1.0, 0.3],
    'P(data|kappa0=-1,kappa=-0.3)': lh['P(data|kappa)'].ix[:, :, -1.0, -0.3]
})
lhr.head()
```

```python
# compute likelihood ratios for the true kappas
lhr['LHR(kappa0=1)'] = (lhr['P(data|kappa0=1,kappa=1)'] / 
			lhr['P(data|kappa0=1,kappa=-1)'])
lhr['LHR(kappa0=-1)'] = (lhr['P(data|kappa0=-1,kappa=-1)'] / 
			lhr['P(data|kappa0=-1,kappa=1)'])
lhr.head()
```

```python
# compute whether the feedback is consistent with model expectations
lhr['consistent|kappa0=1'] = (
    (lhr['LHR(kappa0=1)'] > 1) &
    (lhr['P(data|kappa0=1,kappa=1)'] > 0.5) &
    (lhr['P(data|kappa0=1,kappa=0.3)'] > 0.5))
lhr['consistent|kappa0=-1'] = (
    (lhr['LHR(kappa0=-1)'] > 1) &
    (lhr['P(data|kappa0=-1,kappa=-1)'] > 0.5) &
    (lhr['P(data|kappa0=-1,kappa=-0.3)'] > 0.5))
lhr['consistent'] = lhr['consistent|kappa0=1'] & lhr['consistent|kappa0=-1']
lhr.head()
```

```python
# compute scores based on the likelihood ratios
lhr['score'] = lhr['LHR(kappa0=1)'] * lhr['LHR(kappa0=-1)']
lhr[lhr['consistent']].head()
```

```python
# extract scores for consistent towers, and take the best bitstr for
# each tower
scores = pd.DataFrame(
    lhr[lhr['consistent']]\
    .reset_index()\
    .groupby('base')\
    .apply(lambda x: x.irow(x['score'].argmax()))\
    .set_index(STIMULUS)['score'], columns=['score'])\
    .sort('score', ascending=False)
    
# also include what the feedback is for each tower and kappa0
scores['unstable|kappa0=1'] = fb[:, :, 1.0].ix[scores.index]
scores['unstable|kappa0=-1'] = fb[:, :, -1.0].ix[scores.index]
scores.head()
```

```python
# choose 10 stable and 10 unstable stimuli for the inference task
stable = scores.groupby('unstable|kappa0=1').get_group(0)[:10]
unstable = scores.groupby('unstable|kappa0=1').get_group(1)[1:11]
stims_b = pd.concat([stable, unstable]).sortlevel()
savestims(stims_b, "mass-all", "mass-inference-G-b", "mass_colors")
stims_b
```

```python
# choose 5 stable and 5 unstable stimuli for the prediction+fb task
stable = scores.groupby('unstable|kappa0=1').get_group(0)[10:15]
unstable = scores.groupby('unstable|kappa0=1').get_group(1)[11:16]
stims_a = pd.concat([stable, unstable]).sortlevel()
savestims(stims_a, "mass-all", "mass-inference-G-a", "mass_colors")
stims_a
```

```python
# choose the mass example
mass_example = scores.groupby('unstable|kappa0=1').get_group(1)[:1]
savestims(mass_example, "mass-all", "mass-inference-example-G", "mass_colors")
mass_example
```

# Training and stable/unstable examples

```python
# load the human stability samheight data package and compute human
# responses
human_dp = dpkg.DataPackage.load(
    DATAPATH.joinpath("human/stability_sameheight.dpkg"))
fall = split_stimnames(human_dp.load_resource('experiment.csv').copy())

# we have to flip the 1-7 responses, because we want high values to
# mean unstable
fall['fall? response'] = 1 - ((fall['fall? response'] - 1) / 6.)
# take mean of subject means
hresp = (
    fall.groupby(['base', 'bitstr', 'pid'])['fall? response'].mean()
    .groupby(level=['base', 'bitstr']).mean()
)
```

```python
# load model stability sameheigh tdata
model_dp = dpkg.DataPackage.load(
    DATAPATH.joinpath("model/stability_sameheight_fall.dpkg"))
model = split_stimnames(model_dp.load_resource('model.csv'))
iparam = (0.0, 0.0, 0.0)
fb = model.pivot_table(
    rows=PARAMS + ['base', 'bitstr'],
    cols='sample',
    values='nfell').ix[iparam][0] > 0
fb = fb.sortlevel()
```

```python
sh = pd.DataFrame({'score': hresp, 'unstable': fb})
# rule out stimuli that people predicted incorrectly
sh['score'][(sh['score'] <= 0.5) & sh['unstable']] = np.nan
sh['score'][(sh['score'] >= 0.5) & ~sh['unstable']] = np.nan
sh = sh.dropna()
sh['score'] = np.abs(sh['score'] - 0.5)

# drop stimuli that were already used in the experiment
sh = (sh.sort('score', ascending=False)
      .drop(zip(*stims_a.index)[0], level='base')
      .drop(zip(*stims_b.index)[0], level='base')
      .drop(zip(*mass_example.index)[0], level='base'))
```

```python
# choose the stable example
stable_example = sh.groupby('unstable').get_group(False)[:1]
savestims(stable_example, "stability-sameheight", "mass-inference-stable-example-G", "original")
```

```python
# choose the unstable example
unstable_example = sh.groupby('unstable').get_group(True)[:1]
savestims(unstable_example, "stability-sameheight", "mass-inference-unstable-example-G", "original")
```

```python
# choose the pretest/posttest stimuli
stable_training = sh.groupby('unstable').get_group(False)[1:4]
unstable_training = sh.groupby('unstable').get_group(True)[1:4]
training = pd.concat([stable_training, unstable_training])
savestims(training, "stability-sameheight", "mass-inference-training-G", "original")
```
